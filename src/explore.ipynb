```python
import pandas as pd

#  CSV en DataFrames
movies_df = pd.read_csv("/workspaces/KNN/data/raw/tmdb_5000_movies.csv")
credits_df = pd.read_csv("/workspaces/KNN/data/raw/tmdb_5000_credits.csv")

# Mostrar las primeras filas de cada DataFrame
print("Primeras filas de movies_df:")
display(movies_df.head())

print("\nPrimeras filas de credits_df:")
display(credits_df.head())
```

```python
# Comprobar columnas 
# Verificar las columnas en movies_df
print("Columnas en movies_df:", movies_df.columns.tolist())

# Verificar las columnas en credits_df
print("Columnas en credits_df:", credits_df.columns.tolist())
```

```python
# Creación de una base de datos y unión de tablas con SQL
import sqlite3

# Crear una conexión a una base de datos en memoria
conn = sqlite3.connect(":memory:")

# Guardar los DataFrames como tablas en SQLite
movies_df.to_sql("movies", conn, index=False, if_exists="replace")
credits_df.to_sql("credits", conn, index=False, if_exists="replace")

# Verificar la estructura de la tabla en SQLite
movies_sql_check = pd.read_sql("PRAGMA table_info(movies);", conn)
credits_sql_check = pd.read_sql("PRAGMA table_info(credits);", conn)

print("Estructura de la tabla 'movies' en SQLite:")
display(movies_sql_check)

print("\nEstructura de la tabla 'credits' en SQLite:")
display(credits_sql_check)

# Modificar la consulta SQL para usar `id` en lugar de `movie_id`
query = """
SELECT 
    m.id AS movie_id, 
    m.title, 
    m.overview, 
    m.genres, 
    m.keywords, 
    c.cast, 
    c.crew
FROM movies m
JOIN credits c ON m.title = c.title
"""

# Crear un nuevo DataFrame con la información unificada
merged_df = pd.read_sql(query, conn)

# Cerrar la conexión a la base de datos
conn.close()

# Mostrar las primeras filas del DataFrame unificado
display(merged_df.head())
```

```python
# Transformación de datos
import ast

# Función para extraer los nombres de una lista de diccionarios en formato JSON
def extract_names(json_str):
    try:
        data = ast.literal_eval(json_str)  # Convertir de string a lista de diccionarios
        return " ".join([item["name"].replace(" ", "") for item in data])
    except:
        return ""

# Aplicar la función a las columnas 'genres' y 'keywords'
merged_df["genres"] = merged_df["genres"].apply(extract_names)
merged_df["keywords"] = merged_df["keywords"].apply(extract_names)

# Mostrar las primeras filas después de la transformación
display(merged_df[["genres", "keywords"]].head())
```

```python
# Transformar la columna cast para obtener los tres primeros actores
import ast

# Función para extraer los nombres de una lista de diccionarios en formato JSON
def extract_names(json_str):
    try:
        data = ast.literal_eval(json_str)  # Convertir de string a lista de diccionarios
        return " ".join([item["name"].replace(" ", "") for item in data])
    except:
        return ""

# Aplicar la función a las columnas 'genres' y 'keywords'
merged_df["genres"] = merged_df["genres"].apply(extract_names)
merged_df["keywords"] = merged_df["keywords"].apply(extract_names)

# Mostrar las primeras filas después de la transformación
display(merged_df[["genres", "keywords"]].head())
```

```python
# Extraer el nombre del director de la columna crew
# Función para extraer el director
def extract_director(json_str):
    try:
        data = ast.literal_eval(json_str)
        for crew_member in data:
            if crew_member["job"] == "Director":
                return crew_member["name"].replace(" ", "")
        return ""
    except:
        return ""

# Aplicar la función
merged_df["crew"] = merged_df["crew"].apply(extract_director)

# Mostrar las primeras filas después de la transformación
display(merged_df[["crew"]].head())
```

```python
# Convertir overview en lista de palabras
# Convertir 'overview' en una lista de palabras
merged_df["overview"] = merged_df["overview"].fillna("").apply(lambda x: x.replace(" ", ""))

# Mostrar las primeras filas después de la transformación
display(merged_df[["overview"]].head())
```

```python
# Crear la columna tags combinando todas las columnas procesadas
merged_df["tags"] = (
    merged_df["overview"] + " " +
    merged_df["genres"] + " " +
    merged_df["keywords"] + " " +
    merged_df["cast"] + " " +
    merged_df["crew"]
)

# Mostrar las primeras filas de la nueva columna
display(merged_df[["tags"]].head())
```

```python
# Construcción del modelo KNN
# Vectorización del Texto
from sklearn.feature_extraction.text import CountVectorizer

# Instanciar el vectorizador
vectorizer = CountVectorizer(max_features=5000, stop_words="english")

# Aplicar transformación a la columna 'tags'
vectors = vectorizer.fit_transform(merged_df["tags"]).toarray()

# Mostrar la forma de la matriz de características
print("Forma de la matriz de características:", vectors.shape)
```

```python
# Calcular la similitud del coseno
from sklearn.metrics.pairwise import cosine_similarity

# Calcular la matriz de similitud del coseno
similarity = cosine_similarity(vectors)

# Mostrar una parte de la matriz de similitud
print("Matriz de similitud:")
display(similarity[:5, :5])
```

```python
# Función de recomendación basada en similitud del coseno
def recommend(movie):
    try:
        # Obtener el índice de la película
        movie_index = merged_df[merged_df["title"] == movie].index[0]

        # Obtener las distancias de similitud
        distances = similarity[movie_index]

        # Ordenar las películas por similitud y obtener las 5 más cercanas
        movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

        # Imprimir los títulos recomendados
        print("Películas recomendadas:")
        for i in movie_list:
            print(merged_df.iloc[i[0]].title)
    
    except IndexError:
        print("La película no se encuentra en la base de datos.")
```
